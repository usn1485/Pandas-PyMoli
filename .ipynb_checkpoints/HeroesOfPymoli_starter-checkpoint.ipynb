```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(10)
```

## Player Count

* Display the total number of players


```python
#Calculate total unique players
TotalPlayers=purchase_data["SN"].nunique()
```

```python
#Create Dataframe to disply Total no of players
Total_df=pd.DataFrame({"Total Players":[TotalPlayers]})
Total_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculate unique Items Count
unique_Items=purchase_data["Item ID"].value_counts()
unique_Items_count=unique_Items.count()

#Calculate Average Price
average_Price=round(purchase_data["Price"].mean(),2)

#Calculate total Number of purchases made by players.
Num_purchases=purchase_data["Item Name"].count()

#Calculate how much revenue made by gaming company.
Total_revenue=purchase_data["Price"].sum()
```

```python
#Create Purachse Summary dataframe
purchase_summary_df=pd.DataFrame({ "Number of Unique Items" : [unique_Items_count],
                                   "Price":[average_Price],
                                   "Number of Purchases":[Num_purchases],
                                   "Total Revenue":[Total_revenue]})
#format using Map function
purchase_summary_df["Price"]=purchase_summary_df["Price"].map("${:.2f}".format)
purchase_summary_df["Total Revenue"]=purchase_summary_df["Total Revenue"].map("${:,.2f}".format)
purchase_summary_df
```

# Gender Demographics

```python
#Calculate Male,Female and Other Gender count by Grouing all same Gender values together.
GenderCount=purchase_data.groupby(["Gender"])["SN"].nunique()

# Calculate Percentage of players by dividing Players by gender by Total no of players
Players_percentage_byGender=round(GenderCount/TotalPlayers*100,2)

#create a datafremae to hold ouput
Gen_Demo_df=pd.DataFrame({"Total Count":GenderCount,
                        "Percentage of Players":Players_percentage_byGender})

#Sort column To arrange them in Descending order. 
Gen_Demo_df.sort_values('Total Count',ascending=False)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

* Create a summary data frame to hold the results

* Optional: give the displayed data cleaner formatting

* Display the summary data frame

```python
#Create new dataframe for Purchasing analysis.

purchaseSummaryByGender_df=purchase_data.loc[:,["Purchase ID","Gender","Item ID","Price"]]

#Group by Gender 
groupby_obj=purchaseSummaryByGender_df.groupby(["Gender"])

#find out how many purchases are done by gender
Purchase_count=groupby_obj["Purchase ID"].count()

#Find out What is average price purchases made by perticular Gender
ava_PurPrice=round(groupby_obj["Price"].mean(),2)

#Find out how much is the Total price  of purchases made by each gender.
Pur_byGender=groupby_obj["Price"].sum()

#Find out what is the average purchases made by each player
ava_pur_perPerson= round(Pur_byGender/GenderCount,2)
```

```python
#Create Purachse Summary By Gender  dataframe
purchaseSummaryByGender_df=pd.DataFrame({ "Purchase Count" : Purchase_count,
                                             "Average Purchase Price":ava_PurPrice,
                                             "Total Purchase Value":Pur_byGender,
                                             "Avg Total Purchase per Person":ava_pur_perPerson})

#format using Map function
purchaseSummaryByGender_df["Average Purchase Price"]=purchaseSummaryByGender_df["Average Purchase Price"].map("${:.2f}".format)
purchaseSummaryByGender_df["Total Purchase Value"]=purchaseSummaryByGender_df["Total Purchase Value"].map("${:,.2f}".format)
purchaseSummaryByGender_df["Avg Total Purchase per Person"]=purchaseSummaryByGender_df["Avg Total Purchase per Person"].map("${:.2f}".format)

#Display the Purchasing Analysis by Gender Frame.
purchaseSummaryByGender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create Age bins
Age_bins=[0,9,14,19,24,29,34,39,155]

# create group names for bins
group_names=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

# Group by Age 
AgeGroup=purchase_data.groupby(['Age'])

#Create new Data frame to hold Age Demographic calculations
Age_Demo_df=purchase_data.copy()

#Create new Age Group column in the exisitng DataFrame 
Age_Demo_df["Age Group"]=pd.cut(Age_Demo_df['Age'],Age_bins, labels=group_names)

#Group by AgeGroup  to calculate Total count and percentage of players by age group.
Age_groupby_data=Age_Demo_df.groupby(["Age Group"])

#Calculate How many players have purchased the items by each Age group
Players_count_byAgeGrp=Age_groupby_data["SN"].nunique()

#Calculate Percetage of players by dividing Players count for each age grp by total number of players.
PercentageOFPlayers=round(Players_count_byAgeGrp/TotalPlayers*100,2)

#Display Age demographics.
Age_Demo_df=pd.DataFrame({ "Total Count" : Players_count_byAgeGrp,
                           "Percentage of Players":PercentageOFPlayers,
                                                 })

Age_Demo_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Calculate Top Spenders 
# Create group by Name to find out how much purchases are made by each player and how much money they have spent.
gr_Name_obj = purchase_data.groupby("SN")

#Calculate Total price of purchases made by each player
pricePerPerson = gr_Name_obj['Price'].sum()

#Calculate how many game items have purchased by each player
PurchaseCount = gr_Name_obj['Price'].count()

#Calculate On an Average how much each player spending on purchasing items in this game.
AvaSpending = pricePerPerson/PurchaseCount

#Create Dataframe to display Top Spenders demographics.
TopSpenders_DF = pd.DataFrame({"Purchase Count":PurchaseCount,
                             "Average Price Per Person":AvaSpending,
                             "Total Price":pricePerPerson})
SorterdTopSpenders_df = TopSpenders_DF.sort_values("Total Price",ascending=False,inplace=True)

#format using Map function
TopSpenders_DF["Average Price Per Person"] = TopSpenders_DF["Average Price Per Person"].map("${:.2f}".format)
TopSpenders_DF["Total Price"] = TopSpenders_DF["Total Price"].map("${:.2f}".format)
TopSpenders_DF.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Calculate Most Popular Item
Most_Pop_DF = purchase_data.loc[:,["Item ID","Item Name","Price"]]

#Group by item id and item name 
grby_itemID = Most_Pop_DF.groupby(['Item ID','Item Name']) 

#Calculate how many times the items have been purchased
Item_Purchase_Count = grby_itemID['Price'].count()

# How much total revenue generated by each game item
Item_Total_Purchase_Price = grby_itemID['Price'].sum()

# How much is the per item price?
Item_price = Item_Total_Purchase_Price / Item_Purchase_Count

#Display Data frame of Top Popular items. 
Most_Pop_DF = pd.DataFrame({"Purchase Count":Item_Purchase_Count,
                             "Item Price":Item_price,
                             "Total Purchase Value":Item_Total_Purchase_Price})

Sorted_Most_Pop_DF = Most_Pop_DF.sort_values("Purchase Count",ascending=False)

#format using Map function
Sorted_Most_Pop_DF["Item Price"] = Sorted_Most_Pop_DF["Item Price"].map("${:.2f}".format)
Sorted_Most_Pop_DF["Total Purchase Value"] = Sorted_Most_Pop_DF["Total Purchase Value"].map("${:.2f}".format)

Sorted_Most_Pop_DF.head(5)
```
